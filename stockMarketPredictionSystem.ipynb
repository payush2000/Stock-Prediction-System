```
#creating a new data frame with only the close column
apple_data=apple.filter(['Close'])
#converting the dataframe to numpy array
apple_dataset=apple_data.values
#getting the number of rows to train the model on 
apple_training_data_len = math.ceil(len(apple_dataset) * 0.8 ) 
print (apple_training_data_len)
```

```
#scale the data
scaler=MinMaxScaler(feature_range=(0,1))
apple_scaled_data = scaler.fit_transform(apple_dataset)
```

```
#creating the training data set
#create the scaled training  data set
apple_train_data=apple_scaled_data[0:apple_training_data_len,:]
#split the data into x_train and y_train
x_train=[]
y_train=[]
for i in range(60,len(apple_train_data)):
  x_train.append(apple_train_data[i-60:i,0])
  y_train.append(apple_train_data[i,0])
```

```
#convert x_train and y_train to numpy array
x_train,y_train=np.array(x_train),np.array(y_train)
```

```
#reshaping the data
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
#x_train.shape
```

```
#building the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

 
```

```
#compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```
#training the model
model.fit(x_train,y_train,batch_size=1,epochs=1)
```

```
#creating the testing data set 
apple_test_data=apple_scaled_data[apple_training_data_len - 60 : , : ]
#create the dataset x_test and y_test
x_test=[]
y_test= apple_dataset[apple_training_data_len : , : ]
for i in range (60, len(apple_test_data)):
  x_test.append(apple_test_data[i-60:i,0])
```

```
#converting the data to a numpy array
x_test=np.array(x_test)
```

```
#reshaping the dataset
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

```
#get  the model predicted price value
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```
#getting the root mean square error
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```
#plot the data 
train = apple[:apple_training_data_len]
valid = apple[apple_training_data_len : ]
valid['Predictions']=predictions

#visulization
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.plot(train['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Train','val','Predictions'])
plt.show()
```

```
#long term moving average SMA100
SMA100=pd.DataFrame()
SMA100['Adj Close price'] = AAPL['Close'].rolling(window=100).mean()
SMA100

```

```
#visualizing the data
plt.figure(figsize=(16,8))
plt.plot(AAPL['Close'])
plt.plot(SMA30['Adj Close price'],label = 'SMA30')
plt.plot(SMA100['Adj Close price'],label = 'SMA100')
plt.title("Apple stock price")
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.legend()
```

```
 #create a new dataframe to store all the data
 data=pd.DataFrame()
 data['AAPL']=AAPL['Close']
 data['SMA30']=SMA30['Adj Close price']
 data['SMA100']=SMA100['Adj Close price']
 data
```

```
#creating a function to signal when to buy and sell an asset
def buy_sell(data):
  sigPriceBuy=[]
  sigPriceSell=[]
  flag = -1

  for i in range(len(data)):
    if data['SMA30'][i] > data['SMA100'][i]:
      if flag!=1:
        sigPriceBuy.append(data['AAPL'][i])
        sigPriceSell.append(np.nan)
        flag=1  
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    elif data['SMA30'][i] < data['SMA100'][i]:
      if flag!=0:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(data['AAPL'][i])
        flag=0
        
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    else:
      sigPriceBuy.append(np.nan)
      sigPriceSell.append(np.nan)
  return (sigPriceBuy,sigPriceSell) 

```

```
#storing the buy and sell signal 
buy_sell=buy_sell(data)
data['Buy_Signal_Price']=buy_sell[0]
data['Sell_Signal_Price']=buy_sell[1]
```

```
#description : trading strategy using bollinger band 
#gettin the data

AAPL = web.DataReader("AAPL",'yahoo',start='2020-01-01',end='2020-11-13')
```

```
#calculating the SMA , upper band , lower band
period=20
AAPL['SMA']=AAPL['Close'].rolling(window=period).mean()
#calculating the standard deviation
AAPL['STD']=AAPL['Close'].rolling(window=period).std()
#upper and lower band
AAPL['Upper'] = AAPL['SMA'] + (2*AAPL['STD'])
AAPL['Lower'] = AAPL['SMA'] - (2*AAPL['STD'])

#creating a list of colums to keep
column_list = ['Close','SMA','Upper','Lower']
#plotting the data
AAPL[column_list].plot(figsize=(16,8))
plt.title("APPLE BOLLINGER BAND ",alpha=0.5)
plt.ylabel("Price")
```

```
new_df=AAPL[period-1:]
new_df   
```

```
#creating a function to create a buy and sell signal
def get_signal(data):
  buy_signal=[]
  sell_signal=[]
  
  for i in range(len(data['Close'])):
    if data['Close'][i] > data['Upper'][i]:
      buy_signal.append(np.nan)
      sell_signal.append(data['Close'][i])
    elif data['Close'][i] < data['Lower'][i]:
      buy_signal.append(data['Close'][i])
      sell_signal.append(np.nan)
    else: 
      buy_signal.append(np.nan)
      sell_signal.append(np.nan)


  return ( buy_signal,sell_signal)
```

```
#creating new column
new_df['Buy'] = get_signal(new_df)[0]

new_df['Sell'] = get_signal(new_df)[1]
```

```
#plotting the data
column_list = ['Close','SMA','Upper','Lower','Buy','Sell']
#plotting the data
x_axis=new_df.index
new_df[column_list].plot(figsize=(16,8))
plt.scatter(x_axis,new_df['Buy'],color='green',lw=3,label='Buy',marker ='^')
plt.scatter(x_axis,new_df['Sell'],color='red',lw=3,label='Sell',marker ='v')
plt.title("APPLE BOLLINGER BAND ",alpha=0.5)
plt.ylabel("Price")
```
